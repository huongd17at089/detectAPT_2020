# Constant

```
data_path = "/content/drive/My Drive/huong/300k_64_v1.csv"
split_size = 0.2
number_of_tree = 10
```

# Load data

```
import pandas as pd

data = pd.read_csv(data_path)
print(data[:10][:])
data = data.sample(frac=1).reset_index(drop=True)
cols = data.columns.values.tolist()
print(cols)
data = data.drop(cols[0], axis = 1)

y = data[cols[-1]]
X = data.drop([cols[-1]], axis = 1)
```

# Random forest model

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import time

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=split_size)
model = RandomForestClassifier(n_estimators=number_of_tree)
t1 = time.time()
model.fit(X_train,y_train)
t2 = time.time()
t3 = time.time()
y_pred = model.predict(X_test)
t4 = time.time()
```

# Report

```
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
matrix = confusion_matrix(y_test,y_pred)

print(classification_report(y_test,y_pred))


print("\n\n")

print("num of tree :",number_of_tree )
print("\n---detail----\n")


print("---precision--- ")
print("macro : ", precision_score(y_test, y_pred, average='macro'))
print("micro : ", precision_score(y_test, y_pred, average='micro'))
print("weighted : ", precision_score(y_test, y_pred, average='weighted'))
print("each class : ", precision_score(y_test, y_pred, average=None))

print("\n---recall--- ")
print("macro : ", recall_score(y_test, y_pred, average='macro'))
print("micro : ", recall_score(y_test, y_pred, average='micro'))
print("weighted : ", recall_score(y_test, y_pred, average='weighted'))
print("each class : ", recall_score(y_test, y_pred, average=None))

print("\n---f1 score--- ")
print("macro : ", f1_score(y_test, y_pred, average='macro'))
print("micro : ", f1_score(y_test, y_pred, average='micro'))
print("weighted : ", f1_score(y_test, y_pred, average='weighted'))
print("each class : ", f1_score(y_test, y_pred, average=None))


print("\n---Acc score--- ")
print(accuracy_score(y_test, y_pred))
print("\n---FPR score--- ")
print(matrix[1][0]/(matrix[1][0]+matrix[1][1]))
print("\n---matrix--- ")
print(matrix)
print("\n---time--- ")
print("train", t2-t1)
print("test", t4-t3)

```
